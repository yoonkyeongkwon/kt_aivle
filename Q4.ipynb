```python
def solution(k, a, b):
    b.sort(reverse = True)
    a.sort()
    answer = 0
    for v1,v2 in zip(a[:k-1],b[:k-1]):
        cnt = 0
        if v1 < v2:
            answer += v2
            cnt = 1
        else:
            break
    answer += sum(a[cnt:])
        
    return answer

k = 3
a = [1,2,5,4,3]
b = [5,5,6,6,5]
answer = solution(k, a, b)
print(answer)
```
